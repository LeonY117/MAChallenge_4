```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import multivariate_normal
```

```python
x1 = np.linspace(-10.0, 10.0, 100)
x2 = np.linspace(-10.0, 10.0, 100)
```

```python
def normal_distribution (mu, sigma, x):
    return (1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2)))
```

```python
x, y = np.mgrid[-10:10:.05, -10:10:.05]
pos = np.dstack((x, y))
rv = multivariate_normal([0, 0])
# rv = multivariate_normal([0, 0], [[1, 0.5], [0.5, 1]])

cm = plt.cm.get_cmap('jet')
plt.figure(figsize = (20, 20))
plt.scatter(x, y, c=rv.pdf(pos), cmap=cm)
plt.axis('off')
plt.savefig('./LogImages/Gaussian.png', dpi=300)
# plt.contourf(x, y, rv.pdf(pos))
```
